```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics as stats
```

```python
data = np.linspace(0,10,100)
```

```python
np.mean(data)
```

```python
np.median(data)
```

```python
stats.mode(data)
```

```python
np.std(data)
```

```python
y_data = norm.pdf(data)
```

```python
plt.figure(figsize = (10,10))
```

```python
plt.plot(data, y_data)
```

```python
plt.scatter(data, y_data)
```
